```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))


# Constants
for k2_over_k1 in np.logspace(-2,2,5):

    # Sigma_2 values (in radians)
    sigma_1 = np.linspace(0, np.pi/2 - 0.01, 1000)

    # Expression
    y = np.arctan( k2_over_k1 * np.tan(sigma_1))

    # Plot
    plt.plot(sigma_1 * 180 / np.pi, y* 180 / np.pi, label=f'$k_2/k_1 = {k2_over_k1:.2f}$')


plt.xlabel(r'$\sigma_1$ (deg)')
plt.ylabel('Output')
plt.title('Plot of $\\arctan\\left(\\frac{k_2}{k_1} \\tan(\\sigma_2)\\right)$')
plt.grid(True)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
